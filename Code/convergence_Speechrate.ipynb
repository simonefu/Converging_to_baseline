```python
N = 9
n_min_words = 5
size_data = 1
N_conv_min_speaker_list = [2,3,4,5,6,8]
N_speaker = []
average_speech_rate =[]
baseline_speech_average = []
baseline_int_speech_average = []
n_tot_conversation = []

for N_conv_min_speaker in N_conv_min_speaker_list:
    
    name_file = 'speech_rate_minword' + str(n_min_words) + 'std_' + 'size%_'+ str(size_data*100)+ '_' + str(N_conv_min_speaker) + 'conv_speaker_baseline'
    speech_rate_filter = download_data(N,name_file)
    
    conv_dati = pd.read_csv("..\conv_tab.csv")
    conv_dati = conv_dati[conv_dati['id_topic'] != ' UNK']
    conv_dati = conv_dati[conv_dati['id_conv'].isin(speech_rate_filter['id_conv'])]
    
    speakers = list(conv_dati['speaker_A'].values)
    speakers += list(conv_dati['speaker_B'].values)
    c=Counter(speakers) 

    ID_SPEAKER = list()
    COUNT_CONV = list()

    for id_speaker in list(set (speakers)):
        ID_SPEAKER.append(id_speaker)
        COUNT_CONV.append(c[id_speaker])
    
    N_speaker.append(len(ID_SPEAKER) )
    
    mean_speech = speech_rate_filter['m_log_pointw_speech_rate'].median()
    average_speech_rate.append(mean_speech)
    
    mean_speech_baseline = speech_rate_filter['baseline_S'].median()
    baseline_speech_average.append(mean_speech_baseline)
    
    mean_speech_baseline_int = speech_rate_filter['baseline_I'].median()
    baseline_int_speech_average.append(mean_speech_baseline_int)
    
    len_conversation = len(speech_rate_filter)
    n_tot_conversation.append(len_conversation)
    
    
    
data = {'n_conversations': [2,3,4,5,6,8],
         'N_speaker': N_speaker,
         'average_speech_rate': average_speech_rate,
         'baseline_speech_average': baseline_speech_average,
        'baseline_int_speech_average':baseline_int_speech_average,
        'n_tot_conversation' : n_tot_conversation}
df = pd.DataFrame.from_dict(data)
df
    
```
